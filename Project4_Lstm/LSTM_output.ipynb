```python
import pickle
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load tokenizer and model
with open("/content/drive/MyDrive/Project4/NextWordModel/tokenizer.pkl", "rb") as f:
    tokenizer = pickle.load(f)

model = load_model("/content/drive/MyDrive/Project4/NextWordModel/next_word_model.h5")

# Set same max_seq_len as training
MAX_SEQ_LEN = 20
TOP_K = 5

def predict_next_words(seed_text, top_k=TOP_K):
    token_list = tokenizer.texts_to_sequences([seed_text])[0]
    token_list = pad_sequences([token_list], maxlen=MAX_SEQ_LEN-1, padding='pre')
    predictions = model.predict(token_list, verbose=0)[0]

    top_indices = predictions.argsort()[-top_k:][::-1]
    results = [(word, predictions[idx]) for word, idx in tokenizer.word_index.items() if idx in top_indices]
    results = sorted(results, key=lambda x: -x[1])
    return results

# Input from user
seed = input("Enter your seed text: ")
predictions = predict_next_words(seed)

print(f"\nTop suggestions for: \"{seed}\"")
for word, confidence in predictions:
    print(f"- {word} (confidence: {confidence:.4f})")

```
